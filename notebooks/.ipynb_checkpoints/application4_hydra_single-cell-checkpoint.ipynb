```python
%matplotlib inline
import os
import re
import numpy as np
import scipy.io as sio
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from aesthetics import params
import matplotlib.ticker
from itca import GreedySearch
mpl.rcParams.update(params)
```

```python
import pyreadr
X = pyreadr.read_r("../data/application4/Hydra_40pcs.rds")[None]
labels = pyreadr.read_r("../data/application4/Hydra_labels.rds")[None]
y = pyreadr.read_r("../data/application4/Hydra_y.rds")[None]
y = np.array(y).flatten() - 1
X = np.array(X)
print("data shape", X.shape)
print("n_classes={}".format(len(np.unique(y))))
```

```python
# save X and y to a npz file
from itca.datasets import load_hydra_data
```

```python
import itca
itca.datasets.load_hydra_data()
```

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
gs = GreedySearch(class_type="nominal")
gs.search(X, y, lda, verbose=True, early_stop=True)  
```

```python
# LDA combines class type 19 and 30 into one class
# Note that in the manuscript the class starts from 1. 
# We encode the class starting from 0 for the convnience of programming.
print(gs.selected.mapping.inverse)
```

```python
from sklearn.manifold import TSNE
tsnes = []
perplexities = [10, 20, 30, 40, 50, 60]
for perp in perplexities:
    tsne = TSNE(perplexity=perp, random_state=perp).fit_transform(X)
    tsnes.append(tsne)
np.save('../data/application4/hydra_tsne.npy', tsnes, allow_pickle=True)
```

## Table 7 and Figure 9: Visualizations of  the Hydra data using t-SNE with respect to different values of perplexity. 

Table summarized the neigboring cells in Figure 9.

```python
color_palette = sns.color_palette("husl", n_colors=38)
color_palette[15] = '#f77189'
unique_labels = np.unique(y)
fig, axes = plt.subplots(2, 3)
for ind in range(len(tsnes)):
    tsne = tsnes[ind]
    row, col = ind // 3, ind % 3
    ax = axes[row, col]
    for yi in unique_labels:
            ax.scatter(tsne[y==yi, 0], tsne[y==yi, 1], label=yi, color=color_palette[yi])
            if (yi + 1 == 19) or  (yi + 1 == 30):
                pass
            else:
                ax.text(np.mean(tsne[y==yi, 0]), np.mean(tsne[y==yi, 1]), yi + 1)
    ax.set_title("Perp={}".format(perplexities[ind]))
    if col == 0:
        ax.set_ylabel("tSNE2")
    if row == 1:
        ax.set_xlabel("tSNE1")
# plt.legend()
fig.set_size_inches(10, 6)
# plt.legend(ncol=5)
plt.tight_layout()
```

## Figure 10: Heatmaps of the first 40 principal components of the hydra single-cell RNA-seq data.

```python
_, counts = np.unique(y, return_counts=True)
comparsion = 15
tentacle = 18
tentnem = 29
X1 = X[y == tentacle, :]
X2 = X[y == tentnem, :]
X3 = X[y == comparsion, :]
X12 = np.concatenate([X1, X2])
X13 = np.concatenate([X1, X3])
row_colors12 = ['#35ae9c'] * X1.shape[0] + ['#af88f4'] * X2.shape[0]
row_colors13 = ['#35ae9c'] * X1.shape[0] + ['#f77189'] * X3.shape[0]
counts, X1.shape, X2.shape, X3.shape, counts[13]
```

```python
g12 = sns.clustermap(X12, row_colors=row_colors12, col_cluster=False, cmap='coolwarm',
                    xticklabels=[], yticklabels=[], figsize=(2.75, 3.5))
g12.cax.set_visible(False)
```

```python
g13 = sns.clustermap(X13, row_colors=row_colors13, col_cluster=False, cmap='coolwarm',
                    xticklabels=[], yticklabels=[],figsize=(2.75, 3.5))
g13.cax.set_visible(False)
```
